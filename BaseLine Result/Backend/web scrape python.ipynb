```python
import json
import requests
from bs4 import BeautifulSoup
import re

def getNewsData():
    headers = {
        "User-Agent":
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36"
    }
    response = requests.get(
        "https://www.google.com/search?q=delhi+news&gl=us&tbm=nws&num=100", headers=headers
    )
    soup = BeautifulSoup(response.content, "html.parser")
    news_results = []

    for el in soup.select("div.SoaBEf"):
        link = el.find("a")["href"]
        news_response = requests.get(link, headers=headers)
        news_soup = BeautifulSoup(news_response.content, "html.parser")
        content_paragraphs = [p.get_text() for p in news_soup.find_all("p")]
        content = ' '.join(content_paragraphs)
        # Remove any extra whitespaces and newlines
        content = re.sub(r'\s+', ' ', content)
        # Extract only the first 50 words
        content = ' '.join(content.split()[:50])

        news_results.append(
            {
                "link": link,
                "title": el.select_one("div.MBeuO").get_text(),
                "snippet": el.select_one(".GI74Re").get_text(),
                "date": el.select_one(".LfVVr").get_text(),
                "source": el.select_one(".NUnG9d span").get_text(),
                "content": content
            }
        )

    print(json.dumps(news_results, indent=2))

getNewsData()
```

```python
import csv
import json
import requests
from bs4 import BeautifulSoup
import re

def getNewsData():
    headers = {
        "User-Agent":
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36"
    }
    response = requests.get(
        "https://www.google.com/search?q=kerala+news&gl=us&tbm=nws&num=100", headers=headers
    )
    soup = BeautifulSoup(response.content, "html.parser")
    news_results = []

    for el in soup.select("div.SoaBEf"):
        link = el.find("a")["href"]
        news_response = requests.get(link, headers=headers)
        news_soup = BeautifulSoup(news_response.content, "html.parser")
        content_paragraphs = [p.get_text() for p in news_soup.find_all("p")]
        content = ' '.join(content_paragraphs)
        # Remove any extra whitespaces and newlines
        content = re.sub(r'\s+', ' ', content)
        # Extract only the first 50 words
        content = ' '.join(content.split()[:50])

        news_results.append(
            {
                "link": link,
                "title": el.select_one("div.MBeuO").get_text(),
                "snippet": el.select_one(".GI74Re").get_text(),
                "date": el.select_one(".LfVVr").get_text(),
                "source": el.select_one(".NUnG9d span").get_text(),
                "content": content
            }
        )

    # Save to CSV
    with open('news_data3.csv', 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['link', 'title', 'snippet', 'date', 'source', 'content']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for news in news_results:
            writer.writerow(news)

    print("Data saved to news_data.csv")

getNewsData()
```
